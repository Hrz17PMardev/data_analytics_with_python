```python
revenues = [120, 250, 80, 310, 95]
tax_rate = 0.18
discount_rate = 0.10

upper_100 = []
between100_300 = []
revenues
```

```python
for r in revenues:
    final_revenue = r * (1 + tax_rate) * (1 - discount_rate)
    if final_revenue > 100:
        upper_100.append(final_revenue)
    if 100 <= final_revenue <= 300:
        between100_300.append(final_revenue)
    

```

```python
print("Final revenues above 100:", upper_100)
print("Final revenues between 100 and 300:", between100_300)
```

# Parenthesis has priority over multiplication and division, so the operations inside the parentheses are executed first. 

# Part two

```python
total_revenue = 0
for r in revenues:
    total_revenue += r * (1 + tax_rate) * (1 - discount_rate)
print(total_revenue)
```

```python
average_revenue = total_revenue / len(revenues)
print(average_revenue)
```

```python
max_revenue = 0
for r in revenues:
    final_revenue = r * (1 + tax_rate) * (1 - discount_rate)
    if final_revenue > max_revenue:
        max_revenue = final_revenue
print(max_revenue)
```

```python
upper_150_count = 0
for r in revenues:
    final_revenue = r * (1 + tax_rate) * (1 - discount_rate)
    if final_revenue > 150:
        upper_150_count += 1
print(upper_150_count)
```

```python
commission = 0.05
revenues_with_commission = []
for r in revenues:
    final_revenue = r * (1 + tax_rate) * (1 - discount_rate)
    final_revenue_with_commission = final_revenue * (1 + commission)
    revenues_with_commission.append(final_revenue_with_commission)
print("Final revenues with commission:", revenues_with_commission)
```
