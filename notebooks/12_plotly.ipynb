# Imports

```python
import numpy as np
import pandas as pd
import plotly.express as px
from utils_plotly import *
```

# Dummy Data

```python
np.random.seed(42)

months = pd.date_range("2024-01-01", periods=6, freq="MS")
regions = ["North", "South", "East"]
products = ["A", "B"]

rows = []

for month in months:
    for region in regions:
        for product in products:
            sales = np.random.randint(80, 220)
            customers = np.random.randint(20, 90)
            units = np.random.randint(10, 70)
            
            rows.append([month, region, product, sales, customers, units])

df_plotly = pd.DataFrame(
    rows,
    columns=["month", "region", "product", "sales", "customers", "units"]
)

df_plotly["month_name"] = df_plotly["month"].dt.strftime("%b")
df_plotly.head()
df_plotly["month"].dtype
```

# Bar Plot

```python
region_sales = (
    df_plotly.groupby("region", as_index=False)["sales"]
             .sum()
)

region_sales
```

# Line Chart

```python
monthly_sales = (
    df_plotly.groupby("month", as_index=False)["sales"]
             .sum()
)
# df_plotly.head()
monthly_sales
```

```python
fig = my_line_chart(
    data=monthly_sales,
    x="month",
    y="sales",
    title="Total Sales by Month",
    x_title="",
    y_title="Total Sales",
    markers=True
)

fig.show()
```

# Hystogram

```python
fig = px.histogram(
    df_plotly,
    x="sales",
    nbins=15,
    title="Distribution of Sales",
)

fig.update_layout(
    xaxis_title="Sales",
    yaxis_title="Count"
)

fig.show()
```

# Scatter Plot

```python
fig = px.scatter(
    df_plotly,
    x="customers",
    y="sales",
    color="region",
    hover_data=["product", "month_name", "units"],
    title="Customers vs Sales",
    size="units",
)

fig.update_layout(
    xaxis_title="Customers",
    yaxis_title="Sales",
    legend_title="Region",
)

fig.show()
```
